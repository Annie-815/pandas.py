```python


# Assuming 'salaries_data' is your DataFrame
# Convert 'BasePay' column to numeric, coerce errors to NaN
salaries_data['BasePay'] = pd.to_numeric(salaries_data['BasePay'], errors='coerce')

# Now, find the maximum BasePay and retrieve corresponding EmployeeName
max_basepay = salaries_data.loc[salaries_data['BasePay'].idxmax()]
print(max_basepay['EmployeeName'])
```

```python
#Find Average base pay of all employee 
salaries_data.groupby('year').mean()['BasePay']
```

```python

average_basepay_by_year = salaries_data.groupby('Year')['BasePay'].mean()
print(average_basepay_by_year)
```

```python
average_basepay_by_JobTitle = salaries_data.groupby('JobTitle')['BasePay'].mean()
print(average_basepay_by_year)
```

```python
# Filter the DataFrame to include only rows where the job title is 'ACCOUNTANT'
accountant_data = salaries_data[salaries_data['JobTitle'] == 'ACCOUNTANT']

# Calculate the mean base pay for employees with the job title 'ACCOUNTANT'
average_basepay_accountant = accountant_data['BasePay'].mean()
print("Average base pay for employees with job title 'ACCOUNTANT':", average_basepay_accountant)
```
